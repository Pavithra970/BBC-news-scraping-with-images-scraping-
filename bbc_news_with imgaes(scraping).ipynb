```python
import nest_asyncio
from crawl4ai import AsyncWebCrawler
import asyncio
import os
import re
import requests
from PIL import Image as PILImage, UnidentifiedImageError
from openpyxl import Workbook
from openpyxl.drawing.image import Image as ExcelImage

nest_asyncio.apply()

# Directories and Constants
IMAGE_DIR = "bbc_images"
os.makedirs(IMAGE_DIR, exist_ok=True)

STANDARD_WIDTH = 100
STANDARD_HEIGHT = 100

def extract_data_using_regex(html, tag, attr=None):
    if tag == "a" and attr == "href":
        return re.findall(r'<a[^>]+href=["\'](.*?)["\'].*?>(.*?)</a>', html)
    elif tag == "img":
        return re.findall(r'<img[^>]+(?:src|data-src)=["\'](.*?)["\']', html)
    else:
        return re.findall(rf'<{tag}[^>]*>(.*?)</{tag}>', html, re.DOTALL)

def download_image(url, save_path):
    try:
        response = requests.get(url, stream=True, timeout=10)
        if response.status_code == 200:
            with open(save_path, "wb") as f:
                for chunk in response.iter_content(1024):
                    f.write(chunk)
            return save_path
    except Exception as e:
        print(f"Error downloading image: {url} - {e}")
    return None

def resize_image(image_path, width, height):
    try:
        with PILImage.open(image_path) as img:
            img = img.resize((width, height))
            img.save(image_path)
    except Exception as e:
        print(f"Error resizing image: {image_path} - {e}")

async def main_bbc():
    async with AsyncWebCrawler() as crawler:
        result = await crawler.arun(url="https://www.bbc.com/news")
        html_content = result.html

        # Extract data using regex
        headings = extract_data_using_regex(html_content, "h2") + extract_data_using_regex(html_content, "h3")
        paragraphs = extract_data_using_regex(html_content, "p")
        links = extract_data_using_regex(html_content, "a", attr="href")
        images = extract_data_using_regex(html_content, "img")

        workbook = Workbook()
        sheet = workbook.active
        sheet.title = "BBC Data"
        sheet.append(["Heading", "Text", "Link Text", "Link URL", "ImagePath", "Image"])

        for i, heading in enumerate(headings):
            text = paragraphs[i] if i < len(paragraphs) else ""
            link_text, link_url = links[i] if i < len(links) else ("", "")
            img_path = ""

            if i < len(images):
                img_url = images[i]
                img_url = img_url if img_url.startswith("http") else f"https://www.bbc.com{img_url}"
                img_filename = f"image_{i + 1}.jpg"
                img_path = os.path.join(IMAGE_DIR, img_filename)

                saved_path = download_image(img_url, img_path)
                if saved_path:
                    resize_image(saved_path, STANDARD_WIDTH, STANDARD_HEIGHT)
                else:
                    img_path = ""

            row = [heading.strip(), text.strip(), link_text.strip(), link_url.strip(), img_path, ""]
            sheet.append(row)

            if img_path and os.path.exists(img_path):
                try:
                    excel_image = ExcelImage(img_path)
                    excel_image.width = STANDARD_WIDTH
                    excel_image.height = STANDARD_HEIGHT
                    img_cell = f"F{sheet.max_row}"
                    sheet.add_image(excel_image, img_cell)
                except Exception as e:
                    print(f"Error adding image to Excel: {img_path} - {e}")

        workbook.save("bbc_data_with_images.xlsx")
        print("Data successfully saved to 'bbc_data_with_images.xlsx'.")

asyncio.run(main_bbc())
```
